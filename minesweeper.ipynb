```python
#importing necessary modules
import random

#function generating minesweeper board with bombs and numbers

def GenerateMineSweeperMap(n, k):
  arr = [[0 for column in range(n)] for row in range(n)]

#loop for inserting bombs at random position and incrementing neighbours by 1 
  for i in range(k):   
    x=random.randint(0, n-1)
    y=random.randint(0, n-1)
    arr[x][y] = "\U0001F4A3"
    #top-left
    if (x >= 1 and x <= n-1) and (y >= 1 and y <= n-1):
      if arr[x-1][y-1] != "\U0001F4A3":
        arr[x-1][y-1] += 1
    #top-right
    if (x>=1 and x<=n-1) and (y>=0 and y<=n-2):
      if arr[x-1][y+1] != "\U0001F4A3":
        arr[x-1][y+1] += 1
    #top-center
    if (x>=1 and x<=n-1) and (y>=0 and y<=n-1):
      if arr[x-1][y] != "\U0001F4A3":
        arr[x-1][y] += 1
    #center-right
    if (x>=0 and x<=n-1) and (y>=0 and y<=n-2):
      if arr[x][y+1] != "\U0001F4A3":
        arr[x][y+1] += 1
    #center-left
    if (x>=0 and x<=n-1) and (y>=1 and y<=n-1):
      if arr[x][y-1] != "\U0001F4A3":
        arr[x][y-1] += 1
    #bottom-left
    if (x>=0 and x<=n-2) and (y>=1 and y<=n-1):
      if arr[x+1][y-1] != "\U0001F4A3":
        arr[x+1][y-1] += 1
    #bottom-right
    if (x>=0 and x<=n-2) and (y>=0 and y<=n-2):
      if arr[x+1][y+1] != "\U0001F4A3":
        arr[x+1][y+1] += 1
    #bottom-center
    if (x>=0 and x<=n-2) and (y>=0 and y<=n-1):
      if arr[x+1][y] != "\U0001F4A3":
        arr[x+1][y] += 1
  return arr
  
```

```python
#function for generating a grid of size as minesweeper map with '-' in all cells
def GeneratePLayerMap(n):
   arr = [["-" for column in range(n)] for row in range(n)]
   return arr
```

```python
#function for displaying the grid to the user
def  DisplayMap(map):
  print("\n")
  for row in map:
     print("    ".join(str(cell) for cell in row))
     print(" ")
```

```python
#function for checking whether all cells are opened or no to continue the game
def CheckPlaying(map):
  for row in map:
    for cell in row:
      if cell == "-":
        return True
  return False
```

```python
#function for displaying the score and asking user whether he wants to continue or no
def CheckContinueGame(name,score):
  print("Player : ",name,"score is : ", score)
  play_again = input("\nDo you wish to play again (YES/NO) : ")
  if play_again.upper() == "NO":
    return False
  return True
```

```python
#main function
def Game():
  playing = True
  print("\n----------------------- WELCOME TO THE GAME ----------------------------")
  name = input("\nEnter your name : ")
  print("\nHii",name,"!!! Hope you're doing great")
  while playing:
    print("\nPlease select the grid size and number of bombs")
    size = int(input("Enter the size of the grid you want to play in : "))
    bomb = int(input("Enter the number of bombs to be placed in a grid : "))
    minesweeper_map = GenerateMineSweeperMap(size, bomb)
    playermap = GeneratePLayerMap(size)
    score = 0
    while(True):
      #checking whether all cells are open or no
      if CheckPlaying(playermap):
        choice = int(input("\n1.Open a cell \n2.Flag a bomb \n "))
        if choice == 1:
          print("\nEnter the cell number that you want to open : ")
          x = int(input("X (0 to n-1) : "))
          y = int(input("Y (0 to n-1) : "))
          if minesweeper_map[x][y] != "\U0001F4A3":
            playermap[x][y] = minesweeper_map[x][y]
            DisplayMap(playermap)
            score += 1
          else:
            print('''You Lost :(
            ~~~~~~~~~~~~~GAME OVER!!!~~~~~~~~~~~~~''')
            DisplayMap(minesweeper_map)
            playing = CheckContinueGame(name, score)
        else: 
          print("\nEnter the cell number that you want to flag : ")
          x = int(input("X (0 to n-1) :"))
          y = int(input("Y (0 to n-1) :"))
          if (minesweeper_map[x][y] == "\U0001F4A3"):
            playermap[x][y] = "\U0001F4A3"
            DisplayMap(playermap)
            print("\n***** Bingooo!!! You have flagged the correct cell *****")
          else:
            print("\nOops!!! You have flagged the wrong cell")

      elif playermap == minesweeper_map:
        DisplayMap(playermap)
        print("\nBINGOO!! YOU HAVE DONE IT!!")
        print("~~~~~~~~~~~~~~~~~~~~~~ YOU WON THE GAME :) ~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        playing = CheckContinueGame(name, score)
        break               #breaking from the inner while loop
      
      else:
        print("YOU LOST :( ") 
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~ GAME OVER ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        DisplayMap(minesweeper_map)
        playing = CheckContinueGame(name,score)
        break
```

```python
#calling main function
Game()
```
